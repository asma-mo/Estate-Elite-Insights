# Calculation health_score

## Clean environmental_health data

### Load data

```python
import pandas as pd

df = pd.read_csv("/content/jeu-de-donnees-deal-flow-2019.csv")
df
```

### Select usefull columns and rename them

Get ride of spaces befor each column

```python
df.columns = df.columns.str.strip()
```

```python
df = df[["COMMUNE", "iOpérationCout"]]
df
```

```python
df = df.rename(columns={
    "COMMUNE" : "city_name",
    "iOpérationCout" : "operation_cost"})
df
```

### check and correct their data type

```python
df.dtypes
```

```python
df["operation_cost"] = df["operation_cost"].str.replace(',', '').astype(float)
```

```python
df.dtypes
```

## Prepare geographical_referential data

```python
geographical_referential = pd.read_csv("/content/clean_geographical_referential.csv")
geographical_referential
```

### choose only neede column

```python
geographical_referential = geographical_referential[["municipality_code","city_name"]]
```

```python
geographical_referential
```

## Perform Merge to add municipality code to environmental_health data

```python
merged_df =pd.merge(df, geographical_referential, on='city_name', how='left')
merged_df
```

### check how many null rows we have

```python
merged_df.isnull().sum()
```

### Null data means --> our added dataframe environmental_health contain cities that isn't listed in Prello's data

so it’s need to be droped

```python
cleaned_df = merged_df.dropna()
cleaned_df
```

```python
cleaned_df.isnull().sum()
```

### Get rid of duplication by group them

```python
enviromental_health = cleaned_df.groupby(["municipality_code","city_name"])['operation_cost'].max().reset_index()
enviromental_health
```

### Export the data frame

```python
import pandas as pd
enviromental_health.to_csv("enviromental_health.csv",index=False)
```
