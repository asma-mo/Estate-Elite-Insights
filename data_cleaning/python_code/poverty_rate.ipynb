# combing to columns from two different table on EPCI code and cleaning the data

Load the data

```python
import pandas as pd
poverty = pd.read_csv("/content/poverty_2019.csv")
codes = pd.read_csv("/content/clean_geographical_referential.csv")
```

choose needed columns

```python
poverty = poverty[["EPCI","Taux de pauvreté au seuil de 60 % (en %)"]]
```

```python
codes
```

```python
codes = codes[["city_name","city_name_normalized","municipality_code","latitude",	"longitude","code_ administrative_services"]]
```

renames columns

```python
codes=codes.rename(columns={
    "code_ administrative_services":"EPCI"
})
```

```python
poverty
```

```python
poverty_rate = poverty.rename(columns={
    "Taux de pauvreté au seuil de 60 % (en %)":"poverty_rate"
})
```

Merge the two columns

```python
poverty_rate
```

merge tables on EPCI

```python
poverty_rate = pd.merge(poverty_rate,codes,on="EPCI",how="left")
```

```python
poverty_rate
```

```python
poverty_rate = poverty_rate.drop(columns=("EPCI"))
```

Apply mask to check and see what municipalities are below poverty line

```python
mask = poverty_rate["poverty_rate"]>40
```

```python
poverty_rate[mask]
```

export in csv

```python
poverty_rate.to_csv("poverty_rate.csv",index=False)
```
