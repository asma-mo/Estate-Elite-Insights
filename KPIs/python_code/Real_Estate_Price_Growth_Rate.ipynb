```python
# Importing Libraries
import pandas as pd
import numpy as np
df = pd.read_csv("/content/average_salary_by_municipality (3).csv")
```

```python
# we have 5 years of data
df.year.unique()
```

```python
df.municipality_code.value_counts()
```

```python
#data for 2019
df2019 = df[(df['year'] =="2019-01-01") ]
df2019
```

```python
#data for 2015
df2015 = df[(df['year'] =="2015-01-01") ]
df2015
```

```python
# Merge the data frames based on municipality_code
merged_df = pd.merge(df2015,df2019, left_on='municipality_code', right_on='municipality_code')

merged_df
```

```python
#Real Estate Price Growth Rate
Growth_Rate=((merged_df['avg_net_salary_y']-merged_df['avg_net_salary_x'])/merged_df['avg_net_salary_x'])*100
Growth_Rate=pd.DataFrame(Growth_Rate)
Growth_Rate
```

```python
#  Select the columns we want to keep
merged_df['Growth_Rate']=Growth_Rate
```

```python
merged_df=merged_df[['municipality_code','Growth_Rate']]
merged_df
```

```python
merged_df.to_csv("Real_Estate_Price_Growth_Rate.csv",index=False)
```
