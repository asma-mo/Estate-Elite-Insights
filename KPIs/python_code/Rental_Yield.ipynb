##calculating the KPI

###premises type

```python
pd.value_counts(df_sales.premise_type)
```

####maison = home

```python
mask_df_sales_Maison = df_sales["premise_type"] == "Maison"
mask_df_sales_Maison
```

```python
df_sales_Maison = df_sales[mask_df_sales_Maison]
df_sales_Maison.head()
```

```python
pd.value_counts(df_sales_Maison.premise_type)
```

####Appartement = apartment

```python
mask_df_sales_Appartement = df_sales["premise_type"] == "Appartement"
mask_df_sales_Appartement
```

```python
df_sales_Appartement = df_sales[mask_df_sales_Appartement]
df_sales_Appartement.head()
```

```python
pd.value_counts(df_sales_Appartement.premise_type)
```

####real estate info house

```python
df_restate_Maison = df_restate[["municipality_code","rental_max_house","rental_min_house","rental_med_all"]]
df_restate_Maison
```

```python
df_restate_Maison.rename(columns = {'rental_max_house':'rental_max'}, inplace = True)
```

```python
df_restate_Maison.rename(columns = {'rental_min_house':'rental_min'}, inplace = True)
```

```python
df_restate_Maison.rename(columns = {'rental_med_all':'rental_med'}, inplace = True)
```

```python
df_restate_Maison
```

####real estate info apartment

```python
df_restate_Appartement = df_restate[["municipality_code","rental_max_apartment","rental_min_apartment","rental_med_all"]]
df_restate_Appartement
```

```python
df_restate_Appartement.rename(columns = {'rental_max_apartment':'rental_max'}, inplace = True)
```

```python
df_restate_Appartement.rename(columns = {'rental_min_apartment':'rental_min'}, inplace = True)
```

```python
df_restate_Appartement.rename(columns = {'rental_med_all':'rental_med'}, inplace = True)
```

```python
df_restate_Appartement
```

###joining tables

####house

```python
df_sales_Maison.head(1)
```

```python
df_sales_Maison.shape
```

```python
df_restate_Maison.head(1)
```

```python
df_restate_Maison.shape
```

```python
rental_yield_maison = pd.merge(df_sales_Maison, df_restate_Maison, on="municipality_code", how="left")
rental_yield_maison.head(1)
```

```python
rental_yield_maison.shape
```

####apartment

```python
df_sales_Appartement.head(1)
```

```python
df_sales_Appartement.shape
```

```python
df_restate_Appartement.head(1)
```

```python
df_restate_Appartement.shape
```

```python
rental_yield_Appartement = pd.merge(df_sales_Appartement, df_restate_Appartement, on="municipality_code", how="left")
rental_yield_Appartement.head(1)
```

```python
rental_yield_Appartement.shape
```

###Concatenate Tables

```python
rental_yield_maison.shape
```

```python
rental_yield_Appartement.shape
```

the output should be 3821729 rows

```python
rental_yield = pd.concat([rental_yield_maison,rental_yield_Appartement], axis = 0)
rental_yield
```

```python
rental_yield.shape
```

###measuring rental_yield KPI

```python
rental_yield.dtypes
```

```python
rental_yield["rental_yield_max"] = rental_yield["surface"] * rental_yield["rental_max"]
rental_yield
```

```python
rental_yield["rental_yield_min"] = rental_yield["surface"] * rental_yield["rental_min"]
rental_yield["rental_yield_med"] = rental_yield["surface"] * rental_yield["rental_med"]
rental_yield
```

##cleaning data

```python
rental_yield.isnull().sum()
```

```python
rental_yield["rental_max"].fillna(0, inplace=True)
rental_yield["rental_min"].fillna(0, inplace=True)
rental_yield["rental_med"].fillna(0, inplace=True)
rental_yield["rental_yield_max"].fillna(0, inplace=True)
rental_yield["rental_yield_min"].fillna(0, inplace=True)
rental_yield["rental_yield_med"].fillna(0, inplace=True)
```

```python
rental_yield.isnull().sum()
```
