```python
import pandas as pd
df1=pd.read_csv("/content/POI_tourist_establishments.csv")
df2=pd.read_csv("/content/POI_touristic_sites_by_municipality.csv")
```

```python
df1
```

```python
df2
```

```python
import pandas as pd

# KPI 1: GROUP BY municipality_code, calculate the mean of importance
kpi_1 = df1.groupby('municipality_code')['importance'].mean()

# KPI 2: GROUP BY municipality_code, calculate the mean of importance
kpi_2 = df2.groupby('municipality_code').agg({'importance': 'mean'})
```

```python
print("KPI 1:")
kpi_1*100
```

```python
kpi_1.reset_index()
```

```python
print("KPI 2:")
kpi_2*100
```

```python
kpi_2.reset_index()
```

```python
kpi_1.to_csv("KPI_tourist_rate.csv", index=True)
```
